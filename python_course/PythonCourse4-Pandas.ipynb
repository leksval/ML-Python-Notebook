```python
# Działanie grupowania i agregacji:
f1 = pd.DataFrame({'A': [0,0,0,1,1,2,2]})
f2 = pd.DataFrame({'B': [1,2,3,4,5,6,7]})
f2.groupby(f1['A']).agg('mean')
```

```python
def missing_stats (dataframe, grouping):
    # wymieramy missing
    na_columns  = dataframe.isna().any()
    na_subframe = dataframe.loc[:, na_columns]
    # jeżeli gdziekolwiek nazwa jest bindowana wtedy dopiero przysłania globalną
    
    # własne funkcje agregujące:
    def na (series):  return series.isna().sum()/ len(series) # sprawdza ilość wystąpienia prawdy i ją grupuje
    def zero(series): return (series == 0).sum()/ len(series)
    def both(series): return na(series) + zero(series)
            
    
    grouped = na_subframe.groupby(dataframe[grouping])
    # kryterium grupującym może być orginalny dataframe
    
    aggregated = grouped.agg([na, zero, both]) # 3 własne kryteria agregujące
    totals = na_subframe.agg([na, zero, both]).unstack()
    totals.name = 'Total' # trzeba nadać nazwę dataframeowi
    aggregated = aggregated.append(totals)
    # stack przenosi z poziomu do pionu 
    return aggregated.T                     # T transponuje
```

```python
missing_stats(df, target_column) 
# wielopoziomowość indeksu pozwala na ... doczytać
```

```python
def impute_by_group (dataframe, grouping, 
                     imputed_column, transform_fn):
    reference = (dataframe.groupby(by = grouping)[imputed_column]
                                        .transform(transform_fn))
    return dataframe[imputed_column].fillna(reference)
```
