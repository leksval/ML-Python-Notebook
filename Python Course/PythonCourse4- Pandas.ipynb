```python
# . można dostać się tylko do kolumny, nie do przecięcia

###   expl['ignore']['village'] = True 

# wywali warning bo przy takiej operacji powstanie kopia
# pierwsza operacja expl['ignore'] jest operacją odczytu (nie ma po nich znaku =),
# zapis następuje dopier przy drugich kwadratowych nawiasach
# nie powinno używać się wielu par kwadratowych nawiasów
# zamiast tego używa się:

expl.loc['village', 'ignore'] = True 


# loc indeksuje jak numpy array ( po numerach osi (jak w matematyce pierwsza oś pionowa))
# UWAGA trzeba uważać na kolejność  po przez przypadek może utworzyć się niepotrzebny wiersz i kolumna
#  taoperacja gwarantuje operację na orginalnym obiekcie 
```
